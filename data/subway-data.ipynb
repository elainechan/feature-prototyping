```python
stations = stops[['station_id', 'station_name']] 
stations_dict = stations.to_dict('records')
stations_dict
```

```python
ind = range(2907)
cols = ['source_name', 'target_name']
edges_names = pd.DataFrame(index=ind, columns=cols)
edges_names
```

```python
edges.iloc[0]
```

```python
for i, r in edges.iterrows(): # row in edges
    for s in stations_dict: # station
        if r['source_id'] == s['station_id']:
            edges_names.iat[i, 0] = s['station_name']
        elif r['target_id'] == s['station_id']:
            edges_names.iat[i, 1] = s['station_name']
edges_names
```

```python
# pd.concat([T1.reset_index(drop=True),T2.reset_index(drop=Tru‌​e)], axis=1)
edges_with_names = pd.concat([edges.reset_index(drop=True), edges_names.reset_index(drop=True)], axis=1)
edges_with_names
```
